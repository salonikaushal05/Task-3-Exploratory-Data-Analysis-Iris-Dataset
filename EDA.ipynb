# Task 3: Exploratory Data Analysis (EDA) on Iris Dataset

## Objective

The objective of this task is to perform Exploratory Data Analysis (EDA) on the Iris dataset to understand data distributions, identify patterns, analyze relationships among features, and visualize important insights using statistical methods and graphical representations.

## Import Required Libraries

Import all necessary libraries for data analysis and visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

## Load Iris Dataset

Load the Iris dataset into a Pandas DataFrame.

```python
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

df["species"] = iris.target

df.head()
```

## Dataset Shape

Display the number of rows and columns in the dataset.

```python
df.shape
```

## Dataset Information

Examine the dataset structure, data types, and non-null values.

```python
df.info()
```

## Statistical Summary

Generate descriptive statistics to understand the distribution of numerical features.

```python
df.describe()
```

## Check Missing Values

Verify whether the dataset contains any missing values.

```python
df.isnull().sum()
```

## Species Distribution

Analyze the number of samples available for each species.

```python
df["species"].value_counts()
```

## Species Count Plot

Visualize the distribution of Iris species.

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x="species",
    data=df
)

plt.title("Species Distribution")
plt.show()
```

## Pairplot Analysis

Visualize relationships among all numerical features.

```python
sns.pairplot(
    df,
    hue="species"
)

plt.show()
```

## Correlation Matrix

Calculate correlations among numerical features.

```python
corr_matrix = df.drop("species", axis=1).corr()

corr_matrix
```

## Correlation Heatmap

Visualize feature correlations using a heatmap.

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```

## Histogram Analysis

Analyze feature distributions.

```python
df.hist(
    figsize=(10,8)
)

plt.show()
```

## Box Plot Analysis

Detect outliers and compare feature distributions.

```python
plt.figure(figsize=(10,6))

sns.boxplot(data=df.iloc[:,:4])

plt.title("Feature Distribution")
plt.show()
```

## Scatter Plot Analysis

Study the relationship between Sepal Length and Petal Length.

```python
sns.scatterplot(
    x="sepal length (cm)",
    y="petal length (cm)",
    hue="species",
    data=df
)

plt.show()
```

## Key Findings

Summarize important observations discovered during EDA.

```python
print("""
Key Findings:

1. The dataset contains 150 observations and 5 columns.
2. No missing values were found.
3. Each species contains 50 samples.
4. Petal Length and Petal Width show strong positive correlation.
5. Setosa species is clearly separable from others.
6. Virginica generally has larger petal dimensions.
7. Correlation heatmap reveals strong relationships among petal features.
""")
```

## Conclusion

EDA was successfully performed on the Iris dataset. Various statistical techniques and visualizations were used to identify patterns, understand feature relationships, and gain insights into species characteristics.
